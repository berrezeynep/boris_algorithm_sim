```python
# Bu satr grafiin satrn hemen altnda kmasn salar
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. Veriyi Oku ---
file_name = 'particle_trajectory2.dat'

try:
    # Veriyi ykle
    data = np.loadtxt(file_name, delimiter='\t', skiprows=1)
    
    # Stunlar ayr
    t = data[:, 0]
    x = data[:, 1]
    y = data[:, 2]
    z = data[:, 3]

    # --- 2. Grafii iz ---
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Parac iz
    ax.plot(x, y, z, label='Elektron Yrngesi', color='blue', linewidth=2)

    # Grsellik Ayarlar
    ax.set_title('Manyetik Alanda Parack Hareketi (Synchrotron)', fontsize=14)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.legend()
    
    # Grafii gster
    plt.show()

except FileNotFoundError:
    print("HATA: .dat dosyas bulunamad")
```

```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

file_name = 'trajectory_boris2.dat'

try:

    data = np.loadtxt(file_name, delimiter='\t', skiprows=1)

    t = data[:, 0]
    x = data[:, 1]
    y = data[:, 2]
    z = data[:, 3]
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot(x, y, z, label='Elektron Yrngesi', color='blue', linewidth=2)

    ax.set_title('3D Helezon Hareketi (Boris Algoritmas ile Enerji Korumal)', fontsize=14)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_zlim(np.min(z), np.max(z))
    ax.legend()

    plt.show()

except FileNotFoundError:
    print("HATA: .dat dosyas bulunamad")
```
